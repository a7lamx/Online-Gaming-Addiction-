# Feature Selection

```python
corr = data.corr()
corr.head()
```

```python
corr.shape
```

```python
columns = np.full((corr.shape[0],), True, dtype=int)
for i in range(corr.shape[0]):
    for j in range(i+1, corr.shape[0]):
        if corr.iloc[i,j] >= 0.9:
            if columns[j]:
                columns[j] = False
```

```python
selected_columns = data.columns[columns]
selected_columns.shape
```

```python
selected_columns
```

```python
Data_selected_featuers=data[selected_columns]
```

```python
Data_selected_featuers
```

#  Split data

```python
x=Data_selected_featuers['Timestamp']
y=Data_selected_featuers.drop(['Timestamp'],axis=1)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=77)

```

```python

print('len of x_train ',x_train.shape)
print('len of x_test ',x_test.shape)
print('len of y_train ',y_train.shape)
print('len of y_test ',y_test.shape)
```
